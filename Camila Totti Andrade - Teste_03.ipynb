## Questão 2 (1,5 pontos)

Suponhamos que $x$ e $t$ são dois vetores com 10 elementos, $x$ varia de 0 a 10 e $t$ varia de 1 a 15. Dada a expressão,
$$
y = \cos(\sin(x)) + \exp \left ( \frac{1}{t}\right ),
$$

a) Escreva um programa que calcule $y1$ utilizando `for`.

b) Calcute $y2$ diretamente. Compare $y1$ e $y2$ usando a função `np.allclose(y1,y2)`.

*Dica*: Defina $x$ e $t$ usando as funções do python, como `arange()` ou `linspace()`. Ambos os vetores contém o mesmo número de elementos (igual a 10), porém com início e fim diferentes.

```python
x = np.arange (1,11)
print ('x = ', x)
t = np.linspace (1,15,10)
print ('t =', t)
```

```python
#fórmula: cos(sin(x)) + exp (1/t) | cosseno = np.cos (x) | seno = np.sin (x)
#                                 | exponencial = np.exp (x)

y1 = []   # cálculo de y1 com for

for i in range(len(t)):
    y = (np.cos(np.sin(x[i]))) + np.exp (1/t[i])
    y1.append (y)

print (y1)
```

```python
y2 = (np.cos(np.sin(x))) + np.exp (1/t)
print (y2)

np.allclose(y1,y2)
```

## Questão 6 (1,5 pontos)

A matriz $\mathbf{A}$ e o vetor $\mathbf{b}$, são representados como uma matriz 2D um vetor 1D respectivamente: 

$$
\mathbf{A} = \left\lbrack\begin{array}{ccc}
0 & 12 & -1\cr
-1 & -1 & -1\cr
11 & 5 & 5
\end{array}\right\rbrack
$$

$$
\mathbf{b} = \left\lbrack\begin{array}{c}
-2\cr
1\cr
7
\end{array}\right\rbrack
$$

Multiplicar a matriz pelo vetor resulta em um outro vetor $\mathbf{c}$, cujos componentes são definidos pela regra geral: 

$$\mathbf{c}_i = \sum_j\mathbf{A}_{i,j}\mathbf{b}_j$$

a) Defina $\mathbf{A}$ e $\mathbf{b}$ como NumPy arrays, e faça a multiplicação de $\mathbf{A}$ por $\mathbf{b}$ usando a regra acima. Será necessário usar `for`aninhado.

b) Compare o resultado usando a função `dot()`.

   A1,1*b1 + A1,2*b2 + A1,3*b3             | Lembrando:  nome_variável [linha][coluna]  - para retornar valores específicos
   A2,1*b1 + A2,2*b2 + A2,3*b3
   A3,1*b1 + A3,2*b2 + A3,3*b3


   0*-2 + 12*1 + 7*-1 =  5
  -1*-2 + -1*1 + -1*7 = -6
  11*-2 +  5*1 +  5*7 = 18

```python
A = np.array ([[0,12,-1],                 # shape = (3,3)    (linha, coluna)
               [-1,-1,-1],
               [11,5,5]])      
b = np.array ([-2,1,7])             # shape = (3,1)    (linha, coluna)
```

```python
A.shape, b.shape
```

```python
# Calcular o produto escalar usando o for

N,M = A.shape

C = np.zeros(N)

for i in range(N):
    for j in range(M):
            C[i] = C[i] + A[i,j]*b[j]
            
C
```

```python
C = np.dot(A,b)
print(C)
```
