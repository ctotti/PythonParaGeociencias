## Quest√£o 1 (1 ponto)

Crie uma fun√ß√£o que informe se um n√∫mero fornecido pelo usu√°rio √© primo.

```python
def se_primo (n):
    
    '''
    Esta fun√ß√£o retorna se um n√∫mero fornecido pelo usu√°rio √© primo ou n√£o
    
    Input:
    n: int
    
    Output:
    Primo: true or false
    
    '''
    for i in range(2,n):
        if n % i == 0:
            return False

    return True
```

## Quest√£o 2 (1 ponto)

Crie uma fun√ß√£o que converta a temperatura de graus Celcius para Fahrenheit (nome sugerido: celsius2fah). Depois utilize-a em um programa que pe√ßa para o usu√°rio informar a temperatura e retorne o valor convertido. Obs.: O par√¢metro de entrada deve ser um float.

```python
def celsius2fah (C):
    '''
    Essa fun√ßao converte a temperatura fornecida em graus Celsius (C) para graus Fahrenheit (F)
    
    Input:
    Temperatura (¬∞C): float
    
    Output:
    Temperatura (¬∞F): float
    '''
    
    F = (9 * C + (32 * 5)) / 5
    
    return F
```

```python
x = (float(input('Insira a temperatura em graus Celsius ')))
y = celsius2fah(x)

print ('A temperatura em graus Fahrenheit √©', y, '¬∞F')
```

## Quest√£o 3 (1 ponto)

Crie uma fun√ß√£o que converta a temperatura de graus Fahrenheit para Celsius (nome sugerido: fah2celsius). Depois utilize-a em um programa que pe√ßa para o usu√°rio informar a temperatura retorne o valor convertido. Obs.: O par√¢metro de entrada deve ser um float.

```python
def fah2celsius (F):
    '''
    Essa fun√ßao converte a temperatura fornecida em graus Fahrenheit (F) para graus Celsius (C).
    
    Input:
    Temperatura (¬∞C): float
    
    Output:
    Temperatura (¬∞F): float
    '''
    # ùê∂  =(5√ó(ùêπ‚àí32))/9
    C = (5 *(F - 32)) / 9
    
    return C
```

```python
x = (float(input('Insira a temperatura em graus Fahrenheit ')))
y = fah2celsius(x)

print ('A temperatura em graus Celsius √©', y, '¬∞C')
```

Com as listas `g_obs`, `gamma`e `h` fornecidas abaixo, fa√ßa um programa em que o usu√°rio decida se deseja calcular a anomalia Ar-livre ou a anomalia Bouguer e retorne (imprima) o resultado correto. Para isso:

```python
# Œîùëîùê¥ùêø = ùëîùëúùëèùë† ‚àí ùõæ ‚àí (‚àí 0,3086 * ‚Ñé) - ANOMALIA AR LIVRE - FUN√á√ÉO
def anml_arlivre (g_obs, gamma, h):
    """
    Calcula a Anomalia Ar Livre.
    
    Inputs:
    g_obs - array
    gamma - array
    h - array
    
    Outputs:
    anml_livre - array 
"""
    var_arliv = g_obs - gamma - (- 0.3086 * h)
    
    return var_arliv
```

```python
# Œîùëîùê¥ùêø = ùëîùëúùëèùë† ‚àí ùõæ ‚àí (‚àí 0.3086 * ‚Ñé + 2ùúãùê∫ùúå √ó 105 * ‚Ñé) - ANOMALIA BOUGUER - FUN√á√ÉO
# ùê∫ = 6.674184√ó10^‚àí11    |    ùúå = 2670
def anml_bouguer (g_obs, gamma, h):
    """
    Calcula a Anomalia Bouguer.
    
    Inputs:
    G_obs - array 
    gamma - array
    h - array
    
    Outputs:
    anml_bouguer - array
"""
    var_bouguer = g_obs - gamma - (- 0.3086 * h + 2 * 3.1415 * (6.674184*10**-11) * 2670 * 10**5 * h)
    
    return var_bouguer
```

```python
calc_anomalia = input('Para calcular a Anomalia Bouguer, digite "bouguer". Para calcular a anomalia Ar Livre, digite "livre": ')

if calc_anomalia == 'bouguer':  
    bouguer = anml_bouguer(g_obs, gamma, h)
    print ('Gbouguer = ', bouguer)
elif calc_anomalia == 'livre':
    livre = anml_arlivre(g_obs, gamma, h)
    print ('Gar-livre = ', livre)
else:
    print ('Entrada inv√°lida.')
```

O arquivo `temperaturas_agosto.txt` cont√©m as temperaturas m√°xima, m√©dia e m√≠nima observadas durante o m√™s de agosto na cidade do Rio de Janeiro. Fa√ßa um programa que importe o arquivo, normalize cada coluna de dados e plote os dados normalizados e n√£o normalizados. Para tanto, fa√ßa:

* importe o arquivo usando a fun√ß√£o `loadtxt()` vista em aula;
* Para normalizar uma coluna de dados, primeiro calcule a m√©dia e o desvio padr√£o e depois calcule o dado normalizado da seguinte maneira:

$$
dado \space normalizado = \frac{dado - media}{desvio \space padrao}
$$
ou seja, para calcular a temperatura m√°xima normalizada, primeiro calcule a media e o desvio padr√£o do array correspondente √† temperatura m√°xima e depois utilize os resultados na f√≥rmula acima.
* Plote **dois** gr√°ficos. Um com as tr√™s temperaturas **n√£o normalizadas** e outro com as tr√™s temperaturas **normalizadas**. (Use o subplot()).

*Obs.: √â necess√°rio calcular o dado normalizado para cada coluna de dados (temperatura m√°xima, m√©dia e m√≠nima). Para calcular a media e o desvio padr√£o de cada um, pode utilizar as fun√ß√µes mean() e std() do numpy.*

*Dica: Crie uma fun√ß√£o para calcular o dado normalizado e use mean() e std() dentro dessa fun√ß√£o.*

```python
fname = 'C:/Users/camil/Documents/Geografia/Disciplinas/Introdu√ß√£o √† Programa√ß√£o para Geocientistas - Flora/Jupyter notebook/Camila Totti Andrade - Teste_04/temperaturas_agosto.txt'
temp_agosto = np.loadtxt(fname)
```

```python
# Dias do mes, temp maxima, temp media, temp minima
dias = temp_agosto [:,0]
tmax = temp_agosto [:,1]
tmed = temp_agosto [:,2]
tmin = temp_agosto [:,3]
```

```python
def dado_norm (d):
    '''
    Fornece o dado normalizado.
    
    Input:
    d - array
    
    Output:
    dado_norm - array
    '''
    dado_norm = (d - np.mean(d))/ np.std(d)
    
    return dado_norm
```

```python
tmax_norm = dado_norm(tmax)
tmed_norm = dado_norm(tmed)
tmin_norm = dado_norm(tmin)
```

```python

plt.figure(figsize=(12,6))

plt.subplot(2,1,1)
plt.plot (dias, tmax, 'r-', label='temp. m√°xima')
plt.plot (dias, tmed, 'y-', label='temp. m√©dia')
plt.plot (dias, tmin,label='temp. m√≠nima')
plt.title ('Oscila√ß√£o de Temperatura do M√™s de Agosto', fontsize=15)
plt.xlabel ('dias')
plt.ylabel ('Temperatura (¬∞C)')
plt.legend (loc='lower right')

plt.subplot(2,1,2)
plt.plot (dias, tmax_norm, 'r-', label='temp. m√°xima')
plt.plot (dias, tmed_norm, 'y-', label='temp. m√©dia')
plt.plot (dias, tmin_norm,label='temp. m√≠nima')
plt.title ('Oscila√ß√£o de Temperatura Normalizada do M√™s de Agosto', fontsize=15)
plt.xlabel ('dias')
plt.ylabel ('Temperatura (¬∞C)')
plt.legend (loc='lower right')

plt.tight_layout()
plt.show()
```
