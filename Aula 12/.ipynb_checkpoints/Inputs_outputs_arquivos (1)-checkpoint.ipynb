## Abrir vários arquivos e armazenar as informações em variáveis

Vamos agora abrir vários arquivos dentro de um diretório. O nome dos arquivos precisam ser renomeados usando números. isso vai facilitar o python a abrir na ordem correta, como no exemplo a seguir.



```python
# Precisamos abrir uma nova biblioteca

import os
```

```python
# Definir o diretório onde estão os arquivos. repare que agora o diretório deve ir 
# apenas até a pasta onde estão os arquivos (pasta Temperaturas). 
#Depois deve incluir uma barra ('/') como no exemplo abaixo

#path = '/home/florislei/Desktop/files/temperaturas/'

path = 'DIRETORIO/temperaturas/

os.chdir(path)
files = sorted(filter(os.path.isfile, os.listdir('.')))
```

```python
# O resultado é uma lista que contenha os nomes dos arquivos nessa ordem abaixo

files
```

Com um loop, vamos imprimir o nome de cada arquivo

```python
for file in files:
    print(file)
```

```python
# Podemos também definir o diretório do arquivo concatenando 
# duas strings. A string "path" e um ítem na lista "files".
# Por exemplos: Vamos definir o diretório COMPLETO do arquivo 01jan.csv:

path+files[0]
```

```python
# Agora vamos criar um array (ou uma matriz) que contenha todos os dados. Ou seja, 
# um array que contenha os dados de cada arquivo 01jan.csv, 02fev.csv, 03mar.csv, 
# 04abr.csv, 05mai.csv, 06jun.csv, 07jul.csv

dados = []

for file in files:
    dados.append(np.loadtxt(path+file,skiprows=1,delimiter=',', unpack=True))
```

```python
# Cada bloco do array dados é um array que contém os valores de cada arquivo

dados
```

```python
# Se quisermos somente o bloco de dados do arquivo de janeiro: 

dados[0]
```

```python
# Se quisermos somente as temperaturas médias de janeiro

dados[0][:][2]
```

Podemos usar os índices, como usado nos exemplos acima, para calcular coisas. 

Por exemplo. Se quisermos calcular a média de temperaturas do ano todo. Podemos, através de um loop, calcular a média dos valores de temperatura média de cada um dos blocos (ou seja, de cada um dos meses)

```python
# Primeiro criamos um array de zeros com 12 elementos, um para cada mês do ano.
media_anual = np.zeros(12)

# Agora vamos Calcular media para cada mes

for i in range(len(files)):
    media_anual[i] = np.mean(dados[i][:][2]) # dados[i][:][2] = A cada iteração, vamos acessar a 
                                             # coluna de temperaturas médias
```

```python
# Agora temos um array que contém a temperatura média de cada mês!
media_anual
```

Vamos plotar A quantidade de chuvas do ano junto com a temperatura média anual

```python
fig, ax = plt.subplots()
fig.set_size_inches(12,4)

# Plot grafico de barras
ax.bar(periodo,chuva, color='gray')
ax.tick_params(axis='y', labelcolor='gray')
plt.xticks(periodo,rotation=45)
plt.ylabel('Chuva (mm) ',color='gray')

# Criar um novo eixo, sendo "x" o eixo em comum (twinx)
ax2 = ax.twinx()

# Plot do segundo gráfico com novo eixo em y
ax2.plot(periodo[:7],media_anual[:7], color='green')
ax2.tick_params(axis='y', labelcolor='green')
plt.ylabel('Temperatura média (°C) ',color='green')
plt.ylim((0,40))

plt.tight_layout()
plt.show()
```

## Máscara

```python
import numpy as np
```

```python
x = np.array([1, 2, 3, 4, 5, 6, 7, 8])
```

Agora, queremos calcular a soma dos elementos **menores do que 4** e **maiores do que 6**. Uma forma entediante é utilizar um loop, verificar se um número preenche estas condições, e depois adicioná-lo a uma variável. Isto seria assim:


```python
total = 0    # variável acumuladora para acumular a soma

for num in x:
    if (num<4) or (num>6):
        total += num
        
print(total)
```

A mesma tarefa pode ser conseguida utilizando o conceito de máscara do Python. Funciona essencialmente com uma lista de Booleans (Verdadeiro/Falso), que quando aplicada à matriz original devolve os elementos de interesse. Aqui, `True` refere-se aos elementos que satisfazem a condição (menor que 4 e maior que 6 no nosso caso), e `False` refere-se aos elementos que não satisfazem a condição.
Vamos primeiro criar esta máscara manualmente.

```python
mask = [True, True, True, False, False, False, True, True]
```

A seguir, passamos esta máscara (lista de Booleans) ao nosso array usando indexação. Isto irá devolver apenas os elementos que satisfazem a condição.

```python
x[mask]
```

E agora é só calcular a soma:

```python
total = np.sum(x[mask])

total
```

Mas não precisamos criar a lista de Booleans manualmente. Podemos criar a máscara usando apenas as condições da seguinte forma:

```python
mask = (x < 4) | (x > 6)  # O símbolo | substitui o "or" 
```

```python
mask
```

```python
np.sum(x[mask])
```

Vamos agora voltar para o nosso exemplo de temperaturas para o mês de julho. Se quisermos criar um novo array somente com as temperaturas menores que 20° podemos fazer do jeito entediante, usando o `for`:

### Desafio

O arquivo `temperatura_2020_RJ.csv` contém as temperaturas médias do Rio de Janeiro para o ano de 2020. A partir destes dados, vamos criar 4 plots de temperatura, um para cada estação do ano. 

O objetivo é separar, em 4 arrays, as temperaturas para cada estação do ano, sendo

* Primavera: temperaturas ente os dias 23 de setembro a 21 de dezembro;
* Verão: temperaturas ente os dias 21 de dezembro a 21 de março;
* Outono: temperaturas ente os dias 21 de março a 21 de junho;
* Inverno: temperaturas ente os dias 21 de junho a 23 de setembro.

Uma vez criados os arrays para cada estação, plotar em 4 gráficos diferentes.

```python
# Primeiro vamos importar a biblioteca matplotlib.pyplot, 
# caso já não esteja importada

import matplotlib.pyplot as plt
import numpy as np
```

Quando o arquivo contém diferentes tipos de dados, a rotina np.loadtxt não é a melhor opção (veja exemplo [aqui](https://campus.datacamp.com/courses/introduction-to-importing-data-in-python/introduction-and-flat-files-1?ex=13)). Neste caso, use a rotina [np.genfromtxt](https://numpy.org/doc/stable/reference/generated/numpy.genfromtxt.html).

```python
fname = '/home/florislei/Desktop/files/temperatura_2020_RJ.csv'

# Vamos importar cada coluna do arquivo para as variáveis correspondentes

data = np.genfromtxt(fname, dtype=str, delimiter=',', unpack=True, skip_header=1,usecols=0)
temperatura = np.genfromtxt(fname, delimiter=',', unpack=True, skip_header=1,usecols=1)
```

```python
data
```

```python
temperatura
```

Como o arquivo contém as temperaturas para o ano 2020, os arrays possuem muitas linhas!

```python
data.shape
```

Cada linha do array `data` corresponde a `ano-mes-dia-hora:minuto:segundo`. Veja abaixo a primeira linha:

```python
print(data[0])
```

Para separar as temperaturas por estações do ano, vamos nos basear no array das datas. Vamos descobrir os índices onde começa e termina cada estação do ano. Para isso, vamos usar a função `np.where()`.

```python
# Índices para primavera
# Primavera: temperaturas ente os dias 23 de setembro a 21 de dezembro;

primavera_inicio = np.where(data == '2020-09-23-00:00:00')[0][0]
primavera_fim = np.where(data == '2020-12-21-00:00:00')[0][0]
```

```python
primavera_inicio
```

```python
primavera_fim
```

```python
print(data[primavera_inicio])
print(temperatura[primavera_inicio])
```

```python
# Índices para verão
# Verão: temperaturas ente os dias 21 de dezembro a 21 de março;

verao_inicio = np.where(data == '2020-12-21-00:00:00')[0][0]
verao_fim = np.where(data == '2020-03-21-00:00:00')[0][0]

# Índices para outono
# Outono: temperaturas ente os dias 21 de março a 21 de junho;

outono_inicio = np.where(data == '2020-03-21-00:00:00')[0][0]
outono_fim = np.where(data == '2020-06-21-00:00:00')[0][0]

# Índices para inverno
# Inverno: temperaturas ente os dias 21 de junho a 23 de setembro.

inverno_inicio = np.where(data == '2020-06-21-00:00:00')[0][0]
inverno_fim = np.where(data == '2020-09-23-00:00:00')[0][0]
```

```python
T_primavera = (temperatura[primavera_inicio:primavera_fim])
dias_primavera = (data[primavera_inicio:primavera_fim])
```

```python
T_outono = (temperatura[outono_inicio:outono_fim])
dias_outono = (data[outono_inicio:outono_fim])

T_inverno = (temperatura[inverno_inicio:inverno_fim])
dias_inverno = (data[inverno_inicio:inverno_fim])
```

```python
# Para as temperaturas de verão precisamos tomar cuidado já que os valores correspondentes ao 
# verão estão no início e no fim do array. Vamos usar o np.append() para juntar as duas partes, 
# começando pelos valores de dezembo:

T_verao = np.append(temperatura[verao_inicio:],temperatura[:verao_fim])
dias_verao = np.append(data[verao_inicio:],data[:verao_fim])
```

```python
dias_verao
```

Agora vamos plotar!

```python
# Definindo os limites máximos e mínimos do eixo y

tmax = 40
tmin = 10
```

```python
plt.figure(figsize= (20,10))

plt.subplot(2,2,1)
plt.plot(dias_primavera,T_primavera, 'pink')
plt.ylabel('Temperatura (°C)', fontsize=18)
plt.ylim(tmin,tmax)
plt.xticks(['2020-09-23-00:00:00', '2020-10-21-00:00:00', '2020-11-21-00:00:00', '2020-12-21-00:00:00'], 
           ['Setembro', 'Outubro', 'Novembro', 'Dezembro'], fontsize=16)
plt.yticks(fontsize=16)
plt.title('Primavera 2020', fontsize=20)

plt.subplot(2,2,2)
plt.plot(dias_verao,T_verao, 'green')
plt.ylabel('Temperatura (°C)', fontsize=18)
plt.ylim(tmin,tmax)
plt.xticks(['2020-12-21-00:00:00', '2020-01-21-00:00:00', '2020-02-21-00:00:00', '2020-03-21-00:00:00'], 
           ['Dezembro', 'Janeiro', 'Fevereiro', 'Março'], fontsize=16)
plt.yticks(fontsize=16)
plt.title('Verão 2020', fontsize=20)

plt.subplot(2,2,3)
plt.plot(dias_outono,T_outono, 'orange')
plt.ylabel('Temperatura (°C)', fontsize=18)
plt.ylim(tmin,tmax)
plt.xticks(['2020-03-21-00:00:00', '2020-04-23-00:00:00', '2020-05-21-00:00:00', '2020-06-21-00:00:00'], 
           ['Março', 'Abril', 'Maio', 'Junho'], fontsize=16)
plt.yticks(fontsize=16)
plt.title('Outono 2020', fontsize=20)

plt.subplot(2,2,4)
plt.plot(dias_inverno,T_inverno, 'blue')
plt.ylabel('Temperatura (°C)', fontsize=18)
plt.ylim(tmin,tmax)
plt.xticks(['2020-06-21-00:00:00', '2020-07-21-00:00:00', '2020-08-21-00:00:00', '2020-09-23-00:00:00'], 
           ['Junho', 'Julho', 'Agosto', 'Setembro'], fontsize=16)
plt.yticks(fontsize=16)
plt.title('Inverno 2020', fontsize=20)

plt.tight_layout()
plt.show()
```
