Uma vez que não dissemos para fazer mais nada com a saída da função, o output mostra diretamente as informações contidas no arquivo. Neste caso, essa saída são os dados que acabámos de carregar. Vamos guardar o conteúdo do arquivo dentro de uma variável.

```python
fname = 'C:/Users/camil/Documents/Geografia/Disciplinas/Introdução à Programação para Geocientistas - Flora/Jupyter notebook/Aula 12/temp_julho.txt'

temp = np.loadtxt(fname)
```

Automaticamente a variável `temp` guarda dados memória. `temp`é um array multidimensional (matriz) e podemos manipulá-lo.

```python
temp.shape
```

```python
# pega a primeira linha inteira

temp[0]
```

Vamos 'desmembrar' essa matriz em vetores (arrays unidimensionais) e guardar os valores de cada **coluna** em um novo array.

```python
# primeira linha e primeira coluna, retorna o dia 1

temp[0,0]
```

```python
# retorna todas as linhas da primeira coluna

dias = (temp[:,0])
dias
```

```python
temp_max = temp[:,1] # retorna todas as linhas da segunda coluna
temp_med = temp[:,2] # retorna todas as linhas da terceira coluna
temp_min = temp[:,3] # retorna todas as linhas da 4° coluna
```

```python
temp_max
```

Uma vez que organizamos os dados em arrays separados podemos fazer qualquer tipo de cálculos, análises ou plots com essas informações.

```python
print(np.mean(temp_max)) # calcula a média das temperaturas máximas
```

```python
print(np.mean(temp[:,1])) # outra forma de calcular a média da temp. máx
```

```python
# definindo 3 variáveis diferentes de uma vez só
maxval, minval, meanval = np.max(temp_max), np.min(temp_min), np.mean(temp_med)

stdmax,stdmin = np.std(temp_max), np.std(temp_min)
```

```python
print('Temperatura máxima = ', maxval, 'com desvio padrão de', stdmax)
print('Temperatura mínima = ', minval, 'com desvio padrão de', stdmin)
print('Temperatura média = ', meanval)
```

Com as informações, vamos descobrir qual dia do mês tivemos a maior e menor temperatura usando a função `argmax()` e `argmin()` do Numpy.

```python
# a função 'argmax()' encontra o argumento do máx de um vetor (tipo a função índice do Numpy)
# definindo o maior valor entre as temperaturas máximas

n = np.argmax(temp_max)
```

```python
dias [n]
```

```python
m = np.argmin(temp_min)
```

```python
dias [m]
```

### Plotar as temperaturas

```python
plt.figure(figsize=(16,4))

plt.plot(dias, temp_max,'r', label = 'Temp. máxima')
plt.plot(dias, temp_min,'b', label = 'Temp. mínima')
plt.plot(dias, temp_med,'g', label = 'Temp. média')
plt.legend ()

plt.xlabel ('Dias')
plt.ylabel ('Temperatura (°C)')
plt.title ('Temperaturas do mês de Julho')
```

### Plotar as temperaturas e incluir barras de erro

https://matplotlib.org/stable/gallery/lines_bars_and_markers/errorbar_limits_simple.html#sphx-glr-gallery-lines-bars-and-markers-errorbar-limits-simple-py

https://matplotlib.org/1.2.1/examples/pylab_examples/errorbar_demo.html

```python
plt.figure(figsize=(16,4))

plt.plot(dias, temp_max,'g', label = 'Temp. máxima')
plt.legend ()
plt.errorbar (dias, temp_max, yerr = stdmax)

plt.xlabel ('Dias', fontsize = '18')
plt.ylabel ('Temperatura (°C)', fontsize = '18')
plt.title ('Temperaturas do mês de Julho', fontsize = '18')

plt.show() # para não aparecer o output de texto
```

## Abrir vários arquivos e armazenar as informações em variáveis

```python
# facilita a manipulação de arquivos dentro de diretórios

import os
```

```python
path = 'C:/Users/camil/Documents/Geografia/Disciplinas/Introdução à Programação para Geocientistas - Flora/Jupyter notebook/Aula 12/temperaturas/temperaturas/'

#essa linha faz acessar o diretório
os.chdir(path)

# criamos uma pasta com uma lista dos arquivos
files = sorted(filter(os.path.isfile, os.listdir('.')))
```

```python
files
```

```python
# em cada interação 'file' recebe o nome do arquivo que me interessa

for file in files:
    print(file)
```

```python
path+files[0]
```

```python
dados = []
k = 0
for file in files:
    # informo o 'path' + o nome do arquivo
    # quando o arquivo é CSV. precisa usar o 'unpack=True' para descompactar o arquivo
    dados.append(np.loadtxt(path+file,skiprows=1,delimiter=',', unpack=True))
#    k +=1
```

```python
dados
```

```python
dados[0][:][2]
```

```python
media_anual = np.zeros(12) # 12 pois é 1 elemento para cada mês

# Calcular media para cada mes
# vai precisar de um loop

for i in range(len(files)):
    print (i)
    media_anual[i] = np.mean(dados[i][:][2])
```

```python
# os zeros são os meses que ainda não tenho informação

media_anual
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(12,4)

# Plot grafico de barras
ax.bar(periodo,chuva, color='gray')
ax.tick_params(axis='y', labelcolor='gray')
plt.xticks(periodo,rotation=45)
plt.ylabel('Chuva (mm) ',color='gray')

# Criar um novo eixo, sendo "x" o eixo em comum (twinx)
ax2 = ax.twinx()

# Plot do segundo gráfico com novo eixo em y
ax2.plot(periodo[:7],media_anual[:7], color='green')
ax2.tick_params(axis='y', labelcolor='green')
plt.ylabel('Temperatura média (°C) ',color='green')
plt.ylim((0,40))

plt.tight_layout()
plt.show()
```

## Máscara

Funciona essencialmente com uma lista de Booleans (Verdadeiro/Falso), que quando aplicada ao array original devolve os elementos de interesse. Vamos ver com exemplos.

Seja nosso array de temperaturas de julho. Se quisermos criar um novo array somente com as temperaturas menores que 20° podemos fazer do jeito tradicional, usando o `for`:

```python
temp_med
```

```python
temp_l20 = []

for i in range(len(temp_med)):
    if temp_med[i] < 20.:
        temp_l20.append(temp_med[i])
    
```

```python
temp_l20
```

Um jeito mais prático de fazer isso é usando o conceito de máscara do python:

```python
# Mascara

l20 = (temp_med < 20)
```

```python
l20
```

```python
temp_l20_masked = temp_med[l20]
```

```python
temp_l20_masked
```

```python
np.allclose(temp_l20,temp_l20_masked)
```

```python
# AULA REVISÃO - MATERIAL ACRESCENTADO 

import numpy as np
```

```python
x = np.array ([1,2,3,4,5,6,7,8])
```

```python
# queremos calcular a soma dos elementos menores que 4 e maiores que 6
# uma forma entediante de fazer isso é usando o loop


total = 0     # variável acumuladora para acumular a soma

for i in range(len(x)):
    if x[i] < 4 or x[i] > 6:
        print (x[i])
        total = total + x[i]
        
print (total)
```

```python
# uma outra forma de fazer é através do conceito de máscara do Python

mask = [True, True, True, False, False, False, True, True]
```

```python
x_masked = x[mask]
print (x_masked)
```
